## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd

metadata_df = pd.read_csv('Zomato Restaurant names and Metadata.csv')
reviews_df = pd.read_csv('Zomato Restaurant reviews.csv')

df = pd.merge(reviews_df, metadata_df, left_on='Restaurant', right_on='Name', how='left')

print("Data successfully loaded and merged! Shape of df:", df.shape)
print(df.head())
````

````python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'[^\w\s]', '', text)
    text = re.sub(r'\d+', '', text)
    tokens = text.split()

    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return " ".join(tokens)

df['Clean_Review'] = df['Review'].apply(clean_text)

print("Text Cleaning Successful!")
````

### What all manipulations have you done and insights you found?

Manipulations Done:
	1.	Data Integration: Merged the Restaurant metadata and Review datasets using the restaurant name as the common key to create a consolidated view.
	2.	Data Type Conversion (Cost): The 'Cost' column contained string characters and commas (e.g., '1,200'). I used Regex to extract the numeric values and converted them into float data types for mathematical operations.
	3.	Data Type Conversion (Rating): The 'Rating' column had the format 'X/5'. I stripped the '/5' string, handled inconsistent formats, and converted it to numerical variables.
	4.	Missing Value Treatment: Dropped rows where critical variables ('Cost', 'Rating', and 'review') were null, as imputing these might introduce heavy bias into our clustering model.
	5.	Feature Engineering (NLP): Used the TextBlob library to analyze the textual reviews and generated a new continuous numerical column called Sentiment_Score (ranging from -1.0 to 1.0) to quantify customer satisfaction mathematically.
Insights Found: The raw dataset had a mix of text and categorical data. After cleaning, we successfully transformed business metrics (Cost, Rating) and customer feedback (Sentiment) into numerical formats, making the dataset 100% ready for Unsupervised Machine Learning algorithms like K-Means.

## ***5. Hypothesis Testing***

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Hypothetical Statement 1:
•	H0: High-cost and Low-cost restaurants have the same average rating.
•	H1: Average ratings differ significantly based on cost.
•	Test: Independent Two-Sample T-Test (P-value < 0.05 implies reject H0).
Hypothetical Statement 2:
•	H0: There is no correlation between numeric Rating and NLP Sentiment Score.
•	H1: There is a significant linear correlation between them.
•	Test: Pearson Correlation (P-value < 0.05 implies reject H0).
Hypothetical Statement 3:
•	H0: Average cost is the same across the top 3 most popular cuisines.
•	H1: Average cost varies significantly across these cuisines.
•	Test: One-Way ANOVA (P-value < 0.05 implies reject H0).

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

•	Research Hypothesis: High-cost restaurants have a significantly different average rating compared to low-cost restaurants.
•	Null Hypothesis (H0): There is no significant difference in the average ratings of low-cost and high-cost restaurants. (Mean Rating of High Cost = Mean Rating of Low Cost)
•	Alternate Hypothesis (H1): There is a significant difference in the average ratings of low-cost and high-cost restaurants. (Mean Rating of High Cost != Mean Rating of Low Cost)

#### 2. Perform an appropriate statistical test.

````python
import scipy.stats as stats


df_test1 = df.dropna(subset=['Cost', 'Rating'])

median_cost = df_test1['Cost'].median()
low_cost_ratings = df_test1[df_test1['Cost'] <= median_cost]['Rating']
high_cost_ratings = df_test1[df_test1['Cost'] > median_cost]['Rating']


t_stat, p_value = stats.ttest_ind(low_cost_ratings, high_cost_ratings, equal_var=False)

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject Null Hypothesis. There is a significant difference in ratings based on cost.")
else:
    print("Conclusion: Fail to reject Null Hypothesis. Cost does not significantly affect ratings.")
````

##### Which statistical test have you done to obtain P-Value?

I have used the Independent Two-Sample T-Test (specifically Welch's t-test which does not assume equal population variance).

##### Why did you choose the specific statistical test?

Since I am comparing the means of a continuous variable (Rating) across two independent, non-overlapping categorical groups (Low-Cost vs High-Cost restaurants), the Independent T-Test is the most mathematically appropriate method to check if the difference in their means is statistically significant or just random chance.

### Hypothetical Statement - 2

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

•	Research Hypothesis: The numeric Star Rating given by a user is directly correlated to the Sentiment Score derived from their written text review.
•	Null Hypothesis (H0): There is no significant correlation between the Rating and the NLP Sentiment Score.
•	Alternate Hypothesis (H1): There is a significant linear correlation between the Rating and the NLP Sentiment Score.

#### 2. Perform an appropriate statistical test.

````python
df_test2 = df.dropna(subset=['Rating', 'Sentiment_Score'])


corr_coeff, p_value = stats.pearsonr(df_test2['Rating'], df_test2['Sentiment_Score'])

print(f"Pearson Correlation Coefficient: {corr_coeff:.4f}")
print(f"P-Value: {p_value:.4e}")

if p_value < 0.05:
    print("Conclusion: Reject Null Hypothesis. There is a strong statistical correlation between Rating and Sentiment Score.")
else:
    print("Conclusion: Fail to reject Null Hypothesis. No significant correlation found.")
````

##### Which statistical test have you done to obtain P-Value?

I have used the Pearson Correlation Coefficient Test.

##### Why did you choose the specific statistical test?

Both Rating and Sentiment_Score are continuous/ordinal numerical variables. The Pearson Correlation test is the standard statistical method to evaluate the strength and direction of the linear relationship between two continuous numerical variables.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

•	Research Hypothesis: The average dining cost significantly differs across the top 3 most popular cuisines (North Indian, Chinese, and Fast Food).
•	Null Hypothesis (H0): The average cost is exactly the same across the top 3 cuisines.
•	Alternate Hypothesis (H1): At least one of the top 3 cuisines has a significantly different average cost compared to the others.

#### 2. Perform an appropriate statistical test.

````python
top_3_cuisines = df['Cuisines'].value_counts().head(3).index.tolist()

cuisine_1 = df[df['Cuisines'] == top_3_cuisines[0]]['Cost'].dropna()
cuisine_2 = df[df['Cuisines'] == top_3_cuisines[1]]['Cost'].dropna()
cuisine_3 = df[df['Cuisines'] == top_3_cuisines[2]]['Cost'].dropna()

f_stat, p_value = stats.f_oneway(cuisine_1, cuisine_2, cuisine_3)

print(f"Top 3 Cuisines: {top_3_cuisines}")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4e}")

if p_value < 0.05:
    print("Conclusion: Reject Null Hypothesis. Average cost varies significantly between different popular cuisines.")
else:
    print("Conclusion: Fail to reject Null Hypothesis. No significant cost difference between these cuisines.")
````

##### Which statistical test have you done to obtain P-Value?

I have used the One-Way ANOVA (Analysis of Variance) test.

##### Why did you choose the specific statistical test?

When we need to compare the means of a continuous variable (Cost) across three or more independent categorical groups (the 3 different Cuisines), a T-test is no longer sufficient. One-Way ANOVA is explicitly designed to determine if there are any statistically significant differences between the means of three or more independent groups.
